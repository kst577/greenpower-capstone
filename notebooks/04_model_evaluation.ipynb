```python
import sqlite3, pandas as pd, json
from pathlib import Path
ROOT = Path().resolve().parents[0]
con = sqlite3.connect(ROOT/'data'/'greenpower.db')
```

## Feature layer sanity check

```python
daily = pd.read_sql('SELECT * FROM consumption_daily', con)
print('daily rows:', len(daily))
print('mean daily kWh:', round(daily.total_kwh.mean(),1))
print('mean peak-to-average:', round(daily.par.mean(),2))
daily.head()
```

## Model comparison

```python
summary = pd.read_csv(ROOT/'outputs'/'evaluation_summary.csv')
summary
```

## Weather correlation

```python
d = daily.dropna(subset=['temp_c'])
print('r(daily load, temperature) =', round(d[['mean_kw','temp_c']].corr().iloc[0,1],3))
print('r(daily load, heating degree-hours) =', round(d[['total_kwh','hdh']].corr().iloc[0,1],3))
```
